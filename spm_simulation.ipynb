```python
import pybamm

# Create a simple Single Particle Model
model = pybamm.lithium_ion.SPM()

# Use default parameters
sim = pybamm.Simulation(model)

# Run a discharge simulation
sim.solve([0, 3600])

# Plot the result
sim.plot()
```

```python
# Add this cell to label your simulation clearly
print("SPM Simulation complete!")
print("Discharge from 3.8V to 3.2V over 1C discharge")
print("Model: Single Particle Model (SPM)")
print("Ready to commit to GitHub!")
```
